```python
def get_gene_expression(experiment, gene_idx):
    """
    Efficiently process all tables for a given gene's expression data
    
    Parameters:
    -----------
    experiment : SOMAExperiment
        The SOMA experiment object
    gene_idx : int
        Index of the gene to query
        
    Returns:
    --------
    numpy.ndarray, numpy.ndarray
        Two arrays: cell indices and their corresponding expression values
    """
    import numpy as np
    
    # Lists to collect data from all tables
    all_cells = []
    all_expressions = []
    
    # Read the gene's data
    expr = experiment.ms['RNA'].X['data'].read(coords=[[gene_idx]])  # Note the double brackets

    
    # Process each table
    for table in expr.tables():
        print(table)
        # Get cell indices and expression values
        cells = table['soma_dim_0'].to_numpy()
        expressions = table['soma_data'].to_numpy()
        
        all_cells.append(cells)
        all_expressions.append(expressions)
    
    # Combine all chunks efficiently
    if all_cells:
        cell_indices = np.concatenate(all_cells)
        expression_values = np.concatenate(all_expressions)
        
        # Sort by cell index if needed
        sort_idx = np.argsort(cell_indices)
        cell_indices = cell_indices[sort_idx]
        expression_values = expression_values[sort_idx]
        
        return cell_indices, expression_values
    else:
        return np.array([]), np.array([])

# Example usage:
if __name__ == "__main__":
    gene_idx = 5  # Example gene index
    
    cells, expressions = get_gene_expression(experiment, 5)
    
    # Basic statistics
    if len(expressions) > 0:
        print(f"Number of cells: {len(cells)}")
        print(f"Mean expression: {expressions.mean():.4f}")
        print(f"Non-zero expressions: {(expressions != 0).sum()}")
        print(f"Percent non-zero: {((expressions != 0).sum() / len(expressions) * 100):.1f}%")
    else:
        print("No data found for this gene")
```

```python
def get_gene_expression(gene_name):
    # Get gene index from var
    var_df = experiment.ms['RNA'].var.read().concat().to_pandas()
    gene_idx = var_df[var_df['var_id'] == gene_name].index[0]
    
    # Get expression data for this gene
    X = experiment.ms['RNA'].X['data'].read()
    
    # Concatenate all tables
    tables_list = []
    for table in X.tables():
        tables_list.append(table.to_pandas())
    table_df = pd.concat(tables_list)
    
    # Filter for rows where col (gene) index matches our gene
    gene_mask = table_df['soma_dim_1'] == gene_idx
    rows = table_df[gene_mask]['soma_dim_0'].to_numpy()
    vals = table_df[gene_mask]['soma_data'].to_numpy()
    
    # Create dense array of expression values
    expression = np.zeros(X.shape()[0])
    expression[rows] = vals
    
    return expression
```
