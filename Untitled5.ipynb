```python

def concat_anndata_keep_sample(adata1, adata2):
   """Concatenate AnnDatas keeping overlapping genes and sample column"""
   
   # Find overlapping genes 
   common_genes = np.intersect1d(adata1.var_names, adata2.var_names)
   print(f"Found {len(common_genes)} overlapping genes")
   
   # Subset to overlapping genes
   adata1 = adata1[:, common_genes].copy()
   adata2 = adata2[:, common_genes].copy()
   
   # Ensure sample column names match
   if 'Sample' in adata1.obs.columns and 'sample' in adata2.obs.columns:
       adata1.obs['sample'] = adata1.obs['Sample']
       adata1.obs = adata1.obs.drop('Sample', axis=1)
   
   # Add source identifier 
   adata1.obs['data_source'] = 'dataset1'
   adata2.obs['data_source'] = 'dataset2'
   
   # Concatenate
   combined = ad.concat(
       [adata1, adata2],
       merge='same',
       keys=['dataset1', 'dataset2'],
       label='data_source'
   )
   
   print(f"Final shape: {combined.shape}")
   print("Sample column check:", 'sample' in combined.obs.columns)
   
   return combined



def add_metadata_to_anndata(adata, metadata_df):
    """
    Add metadata columns to AnnData object's obs table by matching sample columns
    
    Parameters
    ----------
    adata : AnnData
        AnnData object with 'sample' column in obs
    metadata_df : pd.DataFrame
        Metadata DataFrame with 'sample' column
    """
    # Create mapping dictionaries for each column except 'sample'
    for col in metadata_df.columns:
        if col != 'sample':
            mapping = dict(zip(metadata_df['sample'], metadata_df[col]))
            adata.obs[col] = adata.obs['sample'].map(mapping)
    
    # Print mapping statistics
    print("Mapping summary:")
    for col in metadata_df.columns:
        if col != 'sample':
            n_mapped = adata.obs[col].notna().sum()
            print(f"{col}: {n_mapped} cells mapped ({n_mapped/len(adata.obs)*100:.1f}%)")
    
    return adata


def process_and_visualize_subset(adata, batch_key, min_cells=3, min_genes=200, max_genes=8000, 
                                max_pct_mt=20, ribo_threshold=5):
    """
    Complete preprocessing workflow optimized for snRNA-seq data, using raw counts
    """
    print("Initial shape:", adata.shape)
    
    # Make a copy of the raw data
    adata_raw = adata.copy()
    adata = adata.copy()
    
    # QC Metrics
    adata.var['mt'] = adata.var_names.str.startswith('MT-')
    adata.var['ribo'] = adata.var_names.str.startswith(('RPS', 'RPL'))
    adata.var['hb'] = adata.var_names.str.contains('^HB[^(P)]')
    
    sc.pp.calculate_qc_metrics(adata, 
                              qc_vars=['mt', 'ribo', 'hb'],
                              percent_top=None, 
                              log1p=False, 
                              inplace=True)
    
    # Filtering
    print("\nFiltering cells...")
    adata = adata[adata.obs.n_genes_by_counts < max_genes, :]
    adata = adata[adata.obs.n_genes_by_counts > min_genes, :]
    adata = adata[adata.obs.pct_counts_mt < max_pct_mt, :]
    adata = adata[adata.obs.pct_counts_ribo < ribo_threshold, :]
    
    sc.pp.filter_genes(adata, min_cells=min_cells)
    
    # Normalization (not applied to the raw data)
    print("\nLog transforming...")
    sc.pp.log1p(adata)
    
    # HVG selection
    print("\nIdentifying highly variable genes...")
    sc.pp.highly_variable_genes(adata, 
                               n_top_genes=1500,
                               flavor='seurat_v3')
    
    sc.pl.highly_variable_genes(adata)
    adata = adata[:, adata.var.highly_variable]
    
    # Scale data
    print("\nScaling data...")
    sc.pp.scale(adata, max_value=10)
    
    # PCA
    print("\nRunning PCA...")
    sc.pp.pca(adata, n_comps=50)
    #sc.pl.pca_variance_ratio(adata, n_pcs=50)
    
    # Batch effects
    print("\nChecking batch effects...")
   # sc.pl.pca(adata, color=batch_key)
    
    # Neighbors
    print("\nComputing neighbors...")
    sc.pp.neighbors(adata, n_neighbors=30, n_pcs=50)
    
    # UMAP before harmony
    print("\nRunning UMAP...")
    sc.tl.umap(adata,min_dist=0.3, spread=1.0)
    adata.obsm['X_umap_before'] = adata.obsm['X_umap'].copy()
    
    # Harmony batch correction
    print("\nRunning Harmony...")
    adata_raw = run_harmony_scanpy(adata, batch_key)
    
    # Recompute neighbors & UMAP on raw, Harmony-corrected data
    sc.pp.neighbors(adata_raw, use_rep='X_harmony', n_neighbors=30, n_pcs=50)
    sc.tl.umap(adata_raw,min_dist=0.3, spread=1.0)
    adata.obsm['X_umap_after'] = adata_raw.obsm['X_umap'].copy()
    
    # Clustering
    print("\nClustering...")
    sc.tl.leiden(adata_raw, resolution=0.3)
    sc.tl.dendrogram(adata_raw, groupby='leiden')
    sc.tl.rank_genes_groups(adata_raw, 'leiden', method='wilcoxon')
    
    return adata_raw


def run_harmony_scanpy(adata, batch_key, n_pcs=50):
    """
    Run Harmony batch correction on Scanpy AnnData object
    
    Parameters:
    -----------
    adata : AnnData
        Annotated data matrix with normalized and scaled counts
    batch_key : str
        Column name in adata.obs containing batch information
    n_pcs : int
        Number of principal components to use (default: 50)
    """
    
    # Check if PCA has been computed
    if 'X_pca' not in adata.obsm:
        sc.pp.pca(adata, n_comps=n_pcs)
    
    # Get PCA matrix and batch labels
    X_pca = adata.obsm['X_pca']
    
    # Convert batch labels to dataframe
    df_metadata = pd.DataFrame(adata.obs[batch_key])
    
    # Run Harmony
    ho = harmonypy.harmony.run_harmony(
        data_mat=X_pca,
        meta_data=df_metadata,
        vars_use=[batch_key],
        theta=2,            # Diversity clustering penalty parameter
        max_iter_harmony=25 # Maximum iterations
    )
    
    # Store Harmony embeddings
    adata.obsm['X_harmony'] = ho.Z_corr.T
    
    return adata


def analyze_ribosomal_content(adata, ribo_threshold=5):
    """
    Analyze and filter cells based on ribosomal content

    Parameters
    ----------
    adata : AnnData
        Annotated data matrix
    ribo_threshold : float, optional
        Maximum percentage of ribosomal content allowed (default: 5)

    Returns
    -------
    AnnData
        Filtered data
    """
    # Calculate some statistics
    print("\nRibosomal content statistics:")
    print(f"Mean: {adata.obs['pct_counts_ribo'].mean():.2f}%")
    print(f"Median: {adata.obs['pct_counts_ribo'].median():.2f}%")
    print(f"Std: {adata.obs['pct_counts_ribo'].std():.2f}%")

    # Create visualization
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

    # Histogram
    sns.histplot(data=adata.obs, x='pct_counts_ribo', bins=50, ax=ax1)
    ax1.axvline(ribo_threshold, color='red', linestyle='--', label=f'Threshold ({ribo_threshold}%)')
    ax1.set_title('Distribution of Ribosomal Content')
    ax1.set_xlabel('Ribosomal Content (%)')
    ax1.legend()

    # Violin plot split by sample type if available
    if 'sample_type' in adata.obs.columns:
        sns.violinplot(data=adata.obs, y='pct_counts_ribo', x='sample_type', ax=ax2)
        ax2.set_title('Ribosomal Content by Sample Type')
    else:
        sns.violinplot(data=adata.obs, y='pct_counts_ribo', ax=ax2)
        ax2.set_title('Ribosomal Content Distribution')

    plt.tight_layout()

    # Calculate cells that would be filtered
    n_cells_before = adata.n_obs
    high_ribo = adata.obs['pct_counts_ribo'] > ribo_threshold
    n_filtered = high_ribo.sum()

    print(f"\nFiltering summary (threshold {ribo_threshold}%):")
    print(f"Cells before filtering: {n_cells_before}")
    print(f"Cells with high ribo content: {n_filtered} ({n_filtered/n_cells_before*100:.1f}%)")

    # Return filtered data
    return adata[~high_ribo].copy()

def process_and_visualize(adata, batch_key, min_cells=3, min_genes=200, max_genes=8000, 
                        max_pct_mt=20, target_sum=1e4, ribo_threshold=5):
   """
   Complete preprocessing workflow optimized for snRNA-seq data
   """
   print("Initial shape:", adata.shape)

   
   raw_adata = adata.copy()  # Create complete copy before any processing
   adata.raw = raw_adata    # Store as .raw attribute

    
   # QC Metrics
   adata.var['mt'] = adata.var_names.str.startswith('MT-')
   adata.var['ribo'] = adata.var_names.str.startswith(('RPS', 'RPL'))
   adata.var['hb'] = adata.var_names.str.contains('^HB[^(P)]')
   
   sc.pp.calculate_qc_metrics(adata, 
                            qc_vars=['mt', 'ribo', 'hb'],
                            percent_top=None, 
                            log1p=False, 
                            inplace=True)
   
   # QC Plots
   sc.pl.violin(adata, ['n_genes_by_counts', 'total_counts', 'pct_counts_mt', 'pct_counts_ribo'],
                jitter=0.4, multi_panel=True)
   
   # Ribosomal Analysis
   print("\nAnalyzing ribosomal content...")
   fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
   
   sns.histplot(data=adata.obs, x='pct_counts_ribo', bins=50, ax=ax1)
   ax1.axvline(ribo_threshold, color='red', linestyle='--', label=f'Threshold ({ribo_threshold}%)')
   ax1.set_title('Distribution of Ribosomal Content')
   ax1.set_xlabel('Ribosomal Content (%)')
   ax1.legend()
   
   if 'sample_type' in adata.obs.columns:
       sns.violinplot(data=adata.obs, y='pct_counts_ribo', x='sample_type', ax=ax2)
       ax2.set_title('Ribosomal Content by Sample Type')
   else:
       sns.violinplot(data=adata.obs, y='pct_counts_ribo', ax=ax2)
       ax2.set_title('Ribosomal Content Distribution')
   
   plt.tight_layout()
   
   # Filtering
   print("\nFiltering cells...")
   adata = adata[adata.obs.n_genes_by_counts < max_genes, :]
   adata = adata[adata.obs.n_genes_by_counts > min_genes, :]
   adata = adata[adata.obs.pct_counts_mt < max_pct_mt, :]
   adata = adata[adata.obs.pct_counts_ribo < ribo_threshold, :]
   
   sc.pp.filter_genes(adata, min_cells=min_cells)
   
   # Doublet removal
   #print("\nRemoving doublets...")
  # sc.external.pp.scrublet(adata)
  # adata = adata[adata.obs['predicted_doublet'] == False, :]
   
   # Store raw

   
   # Normalization
   print("\nNormalizing and log transforming...")
   sc.pp.normalize_total(adata, target_sum=target_sum)
   sc.pp.log1p(adata)
   
   # HVG selection
   print("\nIdentifying highly variable genes...")
   sc.pp.highly_variable_genes(adata, 
                             n_top_genes=1500,
                             flavor='seurat_v3',
                             batch_key=batch_key)
   
   sc.pl.highly_variable_genes(adata)
   adata = adata[:, adata.var.highly_variable]
   
   # Scale data
   print("\nScaling data...")
   sc.pp.scale(adata, max_value=10)
   
   # PCA
   print("\nRunning PCA...")
   sc.pp.pca(adata, n_comps=50)
   sc.pl.pca_variance_ratio(adata, n_pcs=50)
   
   # Batch effects
   print("\nChecking batch effects...")
   sc.pl.pca(adata, color=batch_key)
   
   # Neighbors
   print("\nComputing neighbors...")
   sc.pp.neighbors(adata, n_neighbors=30, n_pcs=50)
   
   # UMAP before harmony
   print("\nRunning UMAP...")
   sc.tl.umap(adata,min_dist=0.3, spread=1.0)
   adata.obsm['X_umap_before'] = adata.obsm['X_umap'].copy()
   
   # Harmony batch correction
   print("\nRunning Harmony...")
   adata = run_harmony_scanpy(adata, batch_key)
   
   # Recompute neighbors & UMAP
   sc.pp.neighbors(adata, use_rep='X_harmony', n_neighbors=30, n_pcs=50)
   sc.tl.umap(adata,min_dist=0.3, spread=1.0)
   adata.obsm['X_umap_after'] = adata.obsm['X_umap'].copy()
   
   # Clustering
   print("\nClustering...")
   sc.tl.leiden(adata, resolution=0.3)
   sc.tl.dendrogram(adata, groupby='leiden')
   sc.tl.rank_genes_groups(adata, 'leiden', method='wilcoxon')
   
   return adata



def run_harmony_scanpy(adata, batch_key, n_pcs=50):
    """
    Run Harmony batch correction on Scanpy AnnData object
    
    Parameters:
    -----------
    adata : AnnData
        Annotated data matrix with normalized and scaled counts
    batch_key : str
        Column name in adata.obs containing batch information
    n_pcs : int
        Number of principal components to use (default: 50)
    """
    
    # Check if PCA has been computed
    if 'X_pca' not in adata.obsm:
        sc.pp.pca(adata, n_comps=n_pcs)
    
    # Get PCA matrix and batch labels
    X_pca = adata.obsm['X_pca']
    
    # Convert batch labels to dataframe
    df_metadata = pd.DataFrame(adata.obs[batch_key])
    
    # Run Harmony
    ho = harmonypy.harmony.run_harmony(
        data_mat=X_pca,
        meta_data=df_metadata,
        vars_use=[batch_key],
        theta=2,            # Diversity clustering penalty parameter
        max_iter_harmony=25 # Maximum iterations
    )
    
    # Store Harmony embeddings
    adata.obsm['X_harmony'] = ho.Z_corr.T
    
    return adata

import seaborn as sns
def analyze_ribosomal_content(adata, ribo_threshold=5):
    """
    Analyze and filter cells based on ribosomal content
    
    Parameters
    ----------
    adata : AnnData
        Annotated data matrix
    ribo_threshold : float, optional
        Maximum percentage of ribosomal content allowed (default: 5)
        
    Returns
    -------
    AnnData
        Filtered data
    """
    # Calculate some statistics
    print("\nRibosomal content statistics:")
    print(f"Mean: {adata.obs['pct_counts_ribo'].mean():.2f}%")
    print(f"Median: {adata.obs['pct_counts_ribo'].median():.2f}%")
    print(f"Std: {adata.obs['pct_counts_ribo'].std():.2f}%")
    
    # Create visualization
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
    
    # Histogram
    sns.histplot(data=adata.obs, x='pct_counts_ribo', bins=50, ax=ax1)
    ax1.axvline(ribo_threshold, color='red', linestyle='--', label=f'Threshold ({ribo_threshold}%)')
    ax1.set_title('Distribution of Ribosomal Content')
    ax1.set_xlabel('Ribosomal Content (%)')
    ax1.legend()
    
    # Violin plot split by sample type if available
    if 'sample_type' in adata.obs.columns:
        sns.violinplot(data=adata.obs, y='pct_counts_ribo', x='sample_type', ax=ax2)
        ax2.set_title('Ribosomal Content by Sample Type')
    else:
        sns.violinplot(data=adata.obs, y='pct_counts_ribo', ax=ax2)
        ax2.set_title('Ribosomal Content Distribution')
    
    plt.tight_layout()
    
    # Calculate cells that would be filtered
    n_cells_before = adata.n_obs
    high_ribo = adata.obs['pct_counts_ribo'] > ribo_threshold
    n_filtered = high_ribo.sum()
    
    print(f"\nFiltering summary (threshold {ribo_threshold}%):")
    print(f"Cells before filtering: {n_cells_before}")
    print(f"Cells with high ribo content: {n_filtered} ({n_filtered/n_cells_before*100:.1f}%)")
    
    # Return filtered data
    return adata[~high_ribo].copy()




def map_subset_obs(subset_adata, full_adata):
   """
   Map observations from a subset AnnData onto full AnnData based on cell barcodes
   
   Parameters:
   -----------
   subset_adata : AnnData with metadata to map (smaller subset)
   full_adata : AnnData to map onto (larger dataset)
   """
   
   # Get observations from subset
   for col in subset_adata.obs.columns:
       if col not in full_adata.obs.columns:
           # Create mapping dictionary
           mapping = dict(zip(subset_adata.obs.index, subset_adata.obs[col]))
           # Map values, NA for non-matches
           full_adata.obs[col] = full_adata.obs.index.map(mapping)
   
   print(f"Mapped {len(subset_adata.obs.columns)} columns from subset to full data")
   return full_adata




def predict_cell_types(adata):
    """
    Predict cell types based on core marker expression
    """
    # Calculate module scores for each cell type
    print("Calculating cell type scores...")
    
    for cell_type, markers in core_cell_types["Non_Malignant"].items():
        if isinstance(markers, dict):  # For nested types like Myeloid
            for subtype, subtype_markers in markers.items():
                available_markers = [m for m in subtype_markers if m in adata.var_names]
                if available_markers:
                    sc.tl.score_genes(adata, available_markers, score_name=f'{subtype}_score')
                else:
                    print(f"Warning: No markers found for {subtype}")
        else:
            available_markers = [m for m in markers if m in adata.var_names]
            if available_markers:
                sc.tl.score_genes(adata, available_markers, score_name=f'{cell_type}_score')
            else:
                print(f"Warning: No markers found for {cell_type}")
    
    # Get all score columns
    score_cols = [col for col in adata.obs.columns if col.endswith('_score')]
    
    # Assign cell type based on highest score
    adata.obs['predicted_celltype'] = (adata.obs[score_cols]
                                     .idxmax(axis=1)
                                     .str.replace('_score', ''))
    
    # Print summary
    print("\nCell type distribution:")
    print(adata.obs['predicted_celltype'].value_counts())
    
    return adata



def create_2d_umap(adata, color_options, size=4, opacity=0.7):
    import plotly.express as px
    import plotly.graph_objects as go
    import pandas as pd
    import numpy as np
    
    umap_coords = pd.DataFrame(
        adata.obsm['X_umap'][:, :2],
        columns=['UMAP1', 'UMAP2'],
        index=adata.obs.index
    )
    
    fig = go.Figure()
    traces_per_var = {}
    current_trace = 0
    
    for color_by in color_options:
        if color_by not in adata.obs:
            print(f"Warning: {color_by} not found in adata.obs")
            continue
            
        traces_for_this_var = 0
        color_values = adata.obs[color_by].copy()
        
        # Check if numeric
        is_numeric = pd.api.types.is_numeric_dtype(color_values) and not isinstance(color_values.dtype, pd.CategoricalDtype)
        
        if not is_numeric:
            na_mask = color_values.isna()
            if na_mask.any():
                fig.add_trace(
                    go.Scattergl(
                        x=umap_coords.loc[na_mask, 'UMAP1'],
                        y=umap_coords.loc[na_mask, 'UMAP2'],
                        mode='markers',
                        marker=dict(
                            size=size,
                            color='lightgrey',
                            opacity=opacity
                        ),
                        name='NA',
                        visible=(color_by == color_options[0]),
                        showlegend=(color_by == color_options[0]),
                        legendgroup=color_by
                    )
                )
                traces_for_this_var += 1
            
            non_na_mask = ~na_mask
            if non_na_mask.any():
                categories = color_values[non_na_mask].unique()
                colors = (px.colors.qualitative.Set3 * (len(categories) // len(px.colors.qualitative.Set3) + 1))[:len(categories)]
                
                for idx, category in enumerate(categories):
                    mask = color_values == category
                    fig.add_trace(
                        go.Scattergl(
                            x=umap_coords.loc[mask, 'UMAP1'],
                            y=umap_coords.loc[mask, 'UMAP2'],
                            mode='markers',
                            marker=dict(
                                size=size,
                                color=colors[idx],
                                opacity=opacity
                            ),
                            name=str(category),
                            visible=(color_by == color_options[0]),
                            showlegend=(color_by == color_options[0]),
                            legendgroup=color_by
                        )
                    )
                    traces_for_this_var += 1
        else:
            na_mask = color_values.isna()
            if na_mask.any():
                fig.add_trace(
                    go.Scattergl(
                        x=umap_coords.loc[na_mask, 'UMAP1'],
                        y=umap_coords.loc[na_mask, 'UMAP2'],
                        mode='markers',
                        marker=dict(
                            size=size,
                            color='lightgrey',
                            opacity=opacity
                        ),
                        name='NA',
                        visible=(color_by == color_options[0]),
                        showlegend=True,
                        legendgroup=color_by
                    )
                )
                traces_for_this_var += 1
            
            non_na_mask = ~na_mask
            if non_na_mask.any():
                fig.add_trace(
                    go.Scattergl(
                        x=umap_coords.loc[non_na_mask, 'UMAP1'],
                        y=umap_coords.loc[non_na_mask, 'UMAP2'],
                        mode='markers',
                        marker=dict(
                            size=size,
                            color=color_values[non_na_mask],
                            colorscale='viridis',
                            opacity=opacity,
                            showscale=True,
                            colorbar=dict(title=color_by)
                        ),
                        name=color_by,
                        visible=(color_by == color_options[0]),
                        showlegend=False
                    )
                )
                traces_for_this_var += 1
        
        traces_per_var[color_by] = traces_for_this_var
        current_trace += traces_for_this_var
    
    buttons = []
    current_trace = 0
    
    for color_by in color_options:
        if color_by not in traces_per_var:
            continue
        
        n_traces = sum(traces_per_var.values())
        visible = [False] * n_traces
        
        for i in range(traces_per_var[color_by]):
            visible[current_trace + i] = True
        
        buttons.append(
            dict(
                label=color_by,
                method="update",
                args=[{"visible": visible},
                     {"title": f"UMAP colored by {color_by}"}]
            )
        )
        current_trace += traces_per_var[color_by]
    
    fig.update_layout(
        template='plotly_white',
        paper_bgcolor='white',
        plot_bgcolor='white',
        xaxis=dict(
            title='UMAP1',
            gridcolor='lightgrey',
            showgrid=True,
            zeroline=False,
            showline=True,
            linewidth=1,
            linecolor='black'
        ),
        yaxis=dict(
            title='UMAP2',
            gridcolor='lightgrey',
            showgrid=True,
            zeroline=False,
            showline=True,
            linewidth=1,
            linecolor='black'
        ),
        updatemenus=[
            dict(
                buttons=buttons,
                direction="down",
                showactive=True,
                x=0.1,
                y=1.1,
                xanchor="left",
                yanchor="top",
            )
        ],
        font=dict(
            family="Arial, sans-serif",
            size=12,
            color="black"
        ),
        showlegend=True,
        legend=dict(
            yanchor="top",
            y=0.99,
            xanchor="left",
            x=0.01,
            bgcolor="rgba(255, 255, 255, 0.8)",
            bordercolor="lightgrey",
            borderwidth=1
        ),
        margin=dict(l=50, r=50, t=50, b=50),
        height=800,
        width=1000,
        hovermode='closest'
    )
    
    return fig




    
# Example usage:

# Analyze and filter data
#filtered_adata = analyze_ribosomal_content(adata_processed, ribo_threshold=5)
```

```python

glioblastoma_markers_combined = {
    "Malignant_Cells": {
        "NPC_like": [
            # Combined NPC1 and NPC2 markers
            "DLL3", "DLL1", "SOX4", "TUBB3", "HES6", "TAGLN3", "NEU4", 
            "MARCKSL1", "CD24", "STMN1", "TCF12", "BEX1", "OLIG1", "MAP2", 
            "FXYD6", "PTPRS", "MLLT11", "NPPA", "BCAN", "MEST", "ASCL1", 
            "BTG2", "DCX", "NXPH1", "HN1", "PFN2", "SCG3", "MYT1", "CHD7", 
            "GPR56", "TUBA1A", "PCBP4", "ETV1", "SHD", "TNR", "AMOTL2", 
            "DBN1", "HIP1", "ABAT", "ELAVL4", "LMF1", "GRIK2", "SERINC5", 
            "TSPAN13", "ELMO1", "GLCCI1", "SEZ6L", "LRRN1", "SEZ6", "SOX11",
            "STMN2", "RND3", "HMP19", "MIAT", "NSG1", "DLX6-AS1", "NREP", 
            "FNBP1L", "STMN4", "DLX5", "MAP1B", "RBFOX2", "IGFBPL1", 
            "TMEM161B-AS1", "DPYSL3", "SEPT3", "PKIA", "ATP1B1", "DYNC1I1", 
            "CD200", "SNAP25", "PAK3", "NDRG4", "KIF5A", "UCHL1", "ENO2", 
            "KIF5C", "DDAH2", "TUBB2A", "LBH", "LOC150568", "TCF4", "GNG3", 
            "NFIB", "DPYSL5", "CRABP1", "NFIX", "CEP170", "BLCAP"
        ],
        "AC_like": [
            "CST3", "S100B", "SLC1A3", "HEPN1", "HOPX", "MT3", "SPARCL1", 
            "MLC1", "GFAP", "FABP7", "BCAN", "PON2", "METTL7B", "SPARC", 
            "GATM", "RAMP1", "PMP2", "AQP4", "DBI", "EDNRB", "PTPRZ1", "CLU", 
            "PMP22", "ATP1A2", "S100A16", "HEY1", "PCDHGC3", "TTYH1", "NDRG2", 
            "PRCP", "ATP1B2", "AGT", "PLTP", "GPM6B", "F3", "RAB31", "PPAP2B", 
            "ANXA5", "TSPAN7"
        ],
        "OPC_like": [
            "BCAN", "PLP1", "GPR17", "FIBIN", "LHFPL3", "OLIG1", "PSAT1", 
            "SCRG1", "OMG", "APOD", "SIRT2", "TNR", "THY1", "PHYHIPL", 
            "SOX2-OT", "NKAIN4", "LPPR1", "PTPRZ1", "VCAN", "DBI", "PMP2", 
            "CNP", "TNS3", "LIMA1", "CA10", "PCDHGC3", "CNTN1", "SCD5", 
            "P2RX7", "CADM2", "TTYH1", "FGF12", "TMEM206", "NEU4", "FXYD6", 
            "RNF13", "RTKN", "GPM6B", "LMF1", "ALCAM", "PGRMC1", "HRASLS", 
            "BCAS1", "RAB31", "PLLP", "FABP5", "NLGN3", "SERINC5", "EPB41L2", 
            "GPR37L1"
        ],
        "MES_like": [
            # Combined MES1 and MES2 markers
            "HILPDA", "ADM", "DDIT3", "NDRG1", "HERPUD1", "DNAJB9", "TRIB3", 
            "ENO2", "AKAP12", "SQSTM1", "MT1X", "ATF3", "NAMPT", "NRN1", 
            "SLC2A1", "BNIP3", "LGALS3", "INSIG2", "IGFBP3", "PPP1R15A", 
            "VIM", "PLOD2", "GBE1", "SLC2A3", "FTL", "WARS", "ERO1L", "XPOT", 
            "HSPA5", "GDF15", "ANXA2", "EPAS1", "LDHA", "P4HA1", "SERTAD1", 
            "PFKP", "PGK1", "EGLN3", "SLC6A6", "CA9", "BNIP3L", "RPL21", 
            "TRAM1", "UFM1", "ASNS", "GOLT1B", "ANGPTL4", "SLC39A14", 
            "CDKN1A", "HSPA9", "CHI3L1", "ANXA1", "CD44", "MT2A", "C1S", 
            "EFEMP1", "C1R", "SOD2", "IFITM3", "TIMP1", "SPP1", "A2M", 
            "S100A11", "S100A10", "FN1", "LGALS1", "S100A16", "CLIC1", 
            "MGST1", "RCAN1", "TAGLN2", "NPC2", "SERPING1", "C8orf4", "EMP1", 
            "APOE", "CTSB", "C3", "MT1E", "EMP3", "SERPINA3", "ACTN1", 
            "PRDX6", "IGFBP7", "SERPINE1", "PLP2", "MGP", "CLIC4", "GFPT2", 
            "GSN", "NNMT", "TUBA1C", "GJA1", "TNFRSF1A", "WWTR1"
        ]
    }
}
import os
import plotly.graph_objects as go
import pandas as pd
import numpy as np
import scanpy as sc

def create_cell_type_marker_umap(adata, gene_lists, cell_type, size=3, opacity=0.8):
    """
    Memory-optimized UMAP visualization for specific cell type markers
    """
    # Create output directory if it doesn't exist
    output_dir = "gene_marker_plots"
    os.makedirs(output_dir, exist_ok=True)
    
    print(f"Processing {cell_type}...")
    
    # Get UMAP coordinates
    umap_coords = pd.DataFrame(
        adata.obsm['X_umap'][:, :2],
        columns=['UMAP1', 'UMAP2'],
        index=adata.obs.index
    )
    
    # Filter for available genes
    genes_to_plot = [gene for gene in gene_lists[cell_type] if gene in adata.var_names]
    if not genes_to_plot:
        print(f"Warning: No genes found for {cell_type}")
        return None
    
    print(f"Found {len(genes_to_plot)} genes for {cell_type}")
    
    # Pre-fetch all gene expression values
    genes_df = pd.DataFrame(
        adata[:, genes_to_plot].X.toarray(),
        columns=genes_to_plot,
        index=adata.obs.index
    )
    
    # Create figure
    fig = go.Figure()
    
    # Add traces
    for i, gene in enumerate(genes_to_plot):
        fig.add_trace(
            go.Scattergl(
                x=umap_coords['UMAP1'],
                y=umap_coords['UMAP2'],
                mode='markers',
                marker=dict(
                    size=size,
                    color=genes_df[gene],
                    colorscale='viridis',
                    opacity=opacity,
                    showscale=True,
                    colorbar=dict(
                        title=f"{gene} expression",
                        x=1.15
                    )
                ),
                name=gene,
                visible=(i == 0),
                showlegend=False
            )
        )
    
    # Create buttons
    buttons = [{
        'label': gene,
        'method': "update",
        'args': [
            {"visible": [i == j for j in range(len(genes_to_plot))]},
            {"title": f"UMAP - {cell_type} marker: {gene}"}
        ]
    } for i, gene in enumerate(genes_to_plot)]
    
    # Update layout
    fig.update_layout(
        template='plotly_white',
        title=f"UMAP - {cell_type} markers",
        paper_bgcolor='white',
        plot_bgcolor='white',
        xaxis=dict(
            title='UMAP1',
            gridcolor='lightgrey',
            showgrid=True,
            zeroline=False,
            showline=True,
            linewidth=1,
            linecolor='black',
            domain=[0, 0.85]
        ),
        yaxis=dict(
            title='UMAP2',
            gridcolor='lightgrey',
            showgrid=True,
            zeroline=False,
            showline=True,
            linewidth=1,
            linecolor='black'
        ),
        updatemenus=[dict(
            buttons=buttons,
            direction="down",
            showactive=True,
            x=1.2,
            y=1.1,
            xanchor="left",
            yanchor="top",
        )],
        font=dict(
            family="Arial, sans-serif",
            size=12,
            color="black"
        ),
        margin=dict(l=50, r=150, t=50, b=50),
        height=1000,
        width=1600,
        hovermode='closest'
    )
    
    # Save figure
    output_path = os.path.join(output_dir, f"umap_{cell_type}_markers.html")
    fig.write_html(output_path, include_plotlyjs='cdn', full_html=False)
    print(f"Saved plot for {cell_type} to {output_path}")
    
    return fig

def generate_all_cell_type_plots(adata, cell_types):
    """
    Generate plots for all cell types
    """
    print("Starting plot generation...")
    
    for cell_type, markers in cell_types.items():
        try:
            _ = create_cell_type_marker_umap(adata, cell_types, cell_type)
        except Exception as e:
            print(f"Error processing {cell_type}: {str(e)}")
    
    print("\nPlot generation complete. Files saved in gene_marker_plots/")

# Define cell types
cell_types = {
    "NPC_like": glioblastoma_markers_combined["Malignant_Cells"]["NPC_like"],
    "OPC_like": glioblastoma_markers_combined["Malignant_Cells"]["OPC_like"],
    "AC_like": glioblastoma_markers_combined["Malignant_Cells"]["AC_like"],
    "MES_like": glioblastoma_markers_combined["Malignant_Cells"]["MES_like"]
}

#full_adata = sc.read_h5ad('concat.final.processed.sample.corrected.h5ad')
# Generate all plots
generate_all_cell_type_plots(full_adata, cell_types)
```

```python
import anndata
def subset_and_process(data_obj,leiden):
    subset_adata = anndata.AnnData(X=data_obj.raw.X[data_obj.obs['leiden'] == leiden],
                              obs=data_obj.obs[data_obj.obs['leiden'] == leiden],
                              var=data_obj.raw.var)
    # Save raw state
    subset_adata.raw = subset_adata
    adata = process_and_visualize_subset(subset_adata, 
                            batch_key='sample',
                            min_cells=3,
                            min_genes=200, 
                            max_genes=4000,
                            max_pct_mt=5,
                            ribo_threshold=5)
    return(adata)
def subset_and_process_list(data_obj, leiden_list):
    combined_adata = None
    for leiden in leiden_list:
        subset_adata = anndata.AnnData(X=data_obj.raw.X[data_obj.obs['leiden'] == leiden],
                                      obs=data_obj.obs[data_obj.obs['leiden'] == leiden],
                                      var=data_obj.raw.var)
        subset_adata.raw = subset_adata
        if combined_adata is None:
            combined_adata = subset_adata
        else:
            combined_adata = combined_adata.concatenate(subset_adata)
    return(combined_adata)
    #return process_and_visualize_subset(combined_adata, batch_key='sample', min_cells=3, min_genes=200, max_genes=4000, max_pct_mt=5, ribo_threshold=5)



def predict_cell_types(adata, cell_types):
    """
    Predict T cell subtypes based on marker gene expression
    
    Parameters
    ----------
    adata : AnnData
        Annotated data matrix
    cell_types : dict
        Dictionary of cell types and their marker genes
    
    Returns
    -------
    adata : AnnData
        Updated annotated data matrix with predicted cell types
    """
    print("Calculating T cell subtype scores...")
    
    # Calculate module scores for each cell type
    for cell_type, markers in cell_types.items():
        available_markers = [m for m in markers if m in adata.var_names]
        if available_markers:
            print(f"Computing score for {cell_type} using {len(available_markers)}/{len(markers)} markers")
            sc.tl.score_genes(adata, available_markers, score_name=f'{cell_type}_score', use_raw=True)
        else:
            print(f"Warning: No markers found for {cell_type}")
    
    # Get all score columns
    score_cols = [col for col in adata.obs.columns if col.endswith('_score')]
    
    if not score_cols:
        raise ValueError("No cell type scores could be calculated. Check marker gene names.")
    
    # Create score matrix
    scores_df = adata.obs[score_cols]
    
    # Calculate score statistics for each cell
    max_scores = scores_df.max(axis=1)
    
    # Assign cell types based on highest score
    predicted_types = []
    score_differences = []
    
    for idx, row in scores_df.iterrows():
        max_score = row.max()
        max_type = row.idxmax().replace('_score', '')
        
        # Sort scores in descending order
        sorted_scores = row.sort_values(ascending=False)
        score_diff = sorted_scores.iloc[0] - sorted_scores.iloc[1] if len(sorted_scores) > 1 else float('inf')
        
        predicted_types.append(max_type)
        score_differences.append(score_diff)
    
    # Add predictions to adata
    adata.obs['predicted_celltype'] = predicted_types
    adata.obs['prediction_confidence'] = score_differences
    
    # Print summary statistics
    print("\nCell type distribution:")
    cell_type_counts = adata.obs['predicted_celltype'].value_counts()
    total_cells = len(adata.obs)
    
    for cell_type, count in cell_type_counts.items():
        percentage = (count / total_cells) * 100
        print(f"{cell_type}: {count} cells ({percentage:.1f}%)")
    
    # Calculate average expression of markers for each predicted cell type
    print("\nValidating predictions with marker gene expression...")
    
    marker_expr_by_type = {}
    for cell_type, markers in cell_types.items():
        available_markers = [m for m in markers if m in adata.var_names]
        if available_markers:
            cells_of_type = adata.obs['predicted_celltype'] == cell_type
            if cells_of_type.any():
                mean_expr = adata[cells_of_type, available_markers].X.mean(axis=0)
                marker_expr_by_type[cell_type] = {
                    'marker_means': dict(zip(available_markers, mean_expr)),
                    'cell_count': cells_of_type.sum()
                }
    
    # Add marker expression summary to adata.uns
    adata.uns['marker_expression_summary'] = marker_expr_by_type
    
    return adata

def plot_prediction_summary(adata, cell_types):
    """
    Generate summary plots for cell type predictions
    """
    import matplotlib.pyplot as plt
    import seaborn as sns
    
    # Create violin plot of scores by predicted cell type
    score_cols = [col for col in adata.obs.columns if col.endswith('_score')]
    score_data = adata.obs[score_cols + ['predicted_celltype']].melt(
        id_vars=['predicted_celltype'],
        var_name='score_type',
        value_name='score'
    )
    
    plt.figure(figsize=(15, 8))
    sns.violinplot(data=score_data, x='predicted_celltype', y='score', 
                  hue='score_type', split=True)
    plt.xticks(rotation=45, ha='right')
    plt.title('Distribution of Cell Type Scores by Predicted Type')
    plt.tight_layout()
    
    # Create heatmap of average marker expression by predicted cell type
    marker_expr = []
    for cell_type, markers in cell_types.items():
        available_markers = [m for m in markers if m in adata.var_names]
        if available_markers:
            cells_of_type = adata.obs['predicted_celltype'] == cell_type
            if cells_of_type.any():
                mean_expr = adata[cells_of_type, available_markers].X.mean(axis=0)
                marker_expr.append({
                    'cell_type': cell_type,
                    **dict(zip(available_markers, mean_expr))
                })
    
    expr_df = pd.DataFrame(marker_expr).set_index('cell_type')
    
    plt.figure(figsize=(15, 10))
    sns.heatmap(expr_df, cmap='YlOrRd', center=0)
    plt.title('Average Marker Expression by Predicted Cell Type')
    plt.tight_layout()
    
    return



def create_cell_type_marker_umap(adata, gene_lists, cell_type, size=3, opacity=0.8):
    """
    Memory-optimized UMAP visualization for specific cell type markers with Leiden clusters
    """
    # Create output directory if it doesn't exist
    output_dir = "gene_marker_plots"
    os.makedirs(output_dir, exist_ok=True)
    
    print(f"Processing {cell_type}...")
    
    # Get UMAP coordinates
    umap_coords = pd.DataFrame(
        adata.obsm['X_umap'][:, :2],
        columns=['UMAP1', 'UMAP2'],
        index=adata.obs.index
    )
    
    # Filter for available genes
    genes_to_plot = [gene for gene in gene_lists[cell_type] if gene in adata.var_names]
    if not genes_to_plot:
        print(f"Warning: No genes found for {cell_type}")
        return None
    
    print(f"Found {len(genes_to_plot)} genes for {cell_type}")
    
    # Pre-fetch all gene expression values
    genes_df = pd.DataFrame(
        adata[:, genes_to_plot].X.toarray(),
        columns=genes_to_plot,
        index=adata.obs.index
    )
    
    # Create figure
    fig = go.Figure()
    
    # Add gene expression traces
    for i, gene in enumerate(genes_to_plot):
        fig.add_trace(
            go.Scattergl(
                x=umap_coords['UMAP1'],
                y=umap_coords['UMAP2'],
                mode='markers',
                marker=dict(
                    size=size,
                    color=genes_df[gene],
                    colorscale='viridis',
                    opacity=opacity,
                    showscale=True,
                    colorbar=dict(
                        title=f"{gene} expression",
                        x=1.15
                    )
                ),
                name=gene,
                visible=(i == 0),
                showlegend=False,
                hovertemplate="<b>Cell</b>: %{customdata}<br>" +
                            "<b>UMAP1</b>: %{x:.2f}<br>" +
                            "<b>UMAP2</b>: %{y:.2f}<br>" +
                            f"<b>{gene} expression</b>: %{{marker.color:.2f}}<extra></extra>",
                customdata=adata.obs.index
            )
        )
    
    # Convert Leiden clusters to categorical if they aren't already
    if not pd.api.types.is_categorical_dtype(adata.obs['leiden']):
        adata.obs['leiden'] = adata.obs['leiden'].astype('category')
    
    # Get unique clusters and create a color mapping
    unique_clusters = sorted(adata.obs['leiden'].unique())
    n_clusters = len(unique_clusters)
    
    # Create a discrete colorscale
    colors = px.colors.qualitative.Set3[:n_clusters] if n_clusters <= 12 else px.colors.qualitative.Alphabet
    colors = colors[:n_clusters]  # Ensure we only take as many colors as we need
    
    # Create discrete colorscale for plotly
    colorscale = [[i/(n_clusters-1), color] for i, color in enumerate(colors)]
    
    # Map clusters to numerical values for coloring
    cluster_map = {cluster: i for i, cluster in enumerate(unique_clusters)}
    leiden_values = [cluster_map[cluster] for cluster in adata.obs['leiden']]
    
    # Add Leiden cluster trace
    fig.add_trace(
        go.Scattergl(
            x=umap_coords['UMAP1'],
            y=umap_coords['UMAP2'],
            mode='markers',
            marker=dict(
                size=size,
                color=leiden_values,
                colorscale=colorscale,
                opacity=opacity,
                showscale=True,
                colorbar=dict(
                    title="Leiden clusters",
                    x=1.15,
                    ticktext=unique_clusters,
                    tickvals=list(range(n_clusters)),
                    tickmode="array",
                ),
            ),
            name='Leiden clusters',
            visible=False,
            showlegend=False,
            hovertemplate="<b>Cell</b>: %{customdata[0]}<br>" +
                         "<b>UMAP1</b>: %{x:.2f}<br>" +
                         "<b>UMAP2</b>: %{y:.2f}<br>" +
                         "<b>Cluster</b>: %{customdata[1]}<extra></extra>",
            customdata=list(zip(adata.obs.index, adata.obs['leiden']))
        )
    )
    
    # Create buttons for genes
    gene_buttons = [{
        'label': gene,
        'method': "update",
        'args': [
            {"visible": [i == j for j in range(len(genes_to_plot) + 1)]},  # +1 for leiden
            {"title": f"UMAP - {cell_type} marker: {gene}"}
        ]
    } for i, gene in enumerate(genes_to_plot)]
    
    # Add button for Leiden clusters
    leiden_button = {
        'label': 'Leiden clusters',
        'method': "update",
        'args': [
            {"visible": [False] * len(genes_to_plot) + [True]},
            {"title": f"UMAP - {cell_type} - Leiden clusters"}
        ]
    }
    
    # Combine all buttons
    buttons = gene_buttons + [leiden_button]
    
    # Update layout
    fig.update_layout(
        template='plotly_white',
        title=f"UMAP - {cell_type} markers",
        paper_bgcolor='white',
        plot_bgcolor='white',
        xaxis=dict(
            title='UMAP1',
            gridcolor='lightgrey',
            showgrid=True,
            zeroline=False,
            showline=True,
            linewidth=1,
            linecolor='black',
            domain=[0, 0.85]
        ),
        yaxis=dict(
            title='UMAP2',
            gridcolor='lightgrey',
            showgrid=True,
            zeroline=False,
            showline=True,
            linewidth=1,
            linecolor='black'
        ),
        updatemenus=[dict(
            buttons=buttons,
            direction="down",
            showactive=True,
            x=1.2,
            y=1.1,
            xanchor="left",
            yanchor="top",
        )],
        font=dict(
            family="Arial, sans-serif",
            size=12,
            color="black"
        ),
        margin=dict(l=50, r=150, t=50, b=50),
        height=1000,
        width=1600,
        hovermode='closest'
    )
    
    # Save figure
    output_path = os.path.join(output_dir, f"umap_{cell_type}_markers.html")
    fig.write_html(output_path, include_plotlyjs='cdn', full_html=False)
    print(f"Saved plot for {cell_type} to {output_path}")
    
    return fig

def generate_all_cell_type_plots(adata, cell_types):
    """
    Generate plots for all cell types
    """
    print("Starting plot generation...")
    
    for cell_type, markers in cell_types.items():
        try:
            _ = create_cell_type_marker_umap(adata, cell_types, cell_type)
        except Exception as e:
            print(f"Error processing {cell_type}: {str(e)}")
    
    print("\nPlot generation complete. Files saved in gene_marker_plots/")

```
